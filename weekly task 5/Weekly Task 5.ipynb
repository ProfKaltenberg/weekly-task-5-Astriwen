# Question 1

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a line graph of total population over time for the states: NY, PA, CT, NJ (10 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Specify colors to be used in your code. (5 pts)

```python
#1 
import matplotlib.pyplot as plt #here is the matplotlib package that we will use
import pandas as pd
import numpy as np
path = "/Users/boss/Downloads/ECO-590/weekly-task-5-Astriwen/weekly task 5/"
cities = pd.read_csv(path+ 'us-cities-top-1k-multi-year.csv')


select_cities = cities[(cities['State'] == 'New York') | (cities['State'] == 'Pennsylvania')|
                     (cities['State'] == 'Maryland') |(cities['State'] == 'Delaware')] 
colors = ['red','green','blue','yellow'] # sets colors
store_time_state = select_cities.groupby(['year','State'])['Population'].sum().reset_index()#make data set with state i filtered for and taking the year and then get their population for each year you took 
i=0
for s in store_time_state['State'].unique():
    data = store_time_state[store_time_state['State'] ==s]#puts in data if it the info is for that specific state
    plt.plot(data['year'], data['Population'], label=s, color=colors[i])# can also put all the states in for loop
    i += 1 #iterates color

plt.xticks(range(2014,2019)) #unique didnt work since no 2017

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel('year')
plt.ylabel('Population')
plt.title('Population Over Time')
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

# Question 2

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a Bar Graph of the total population for 5 cities in the state of New York comparing 2014 years and 2018 years (15 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Combine the graph in question 1 and question 2 in one figure.

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,6))


select_cities = cities[(cities['State'] == 'New York') | (cities['State'] == 'Pennsylvania')|
                     (cities['State'] == 'Maryland') |(cities['State'] == 'Delaware')] 
colors = ['red','green','blue','yellow'] # sets colors
store_time_state = select_cities.groupby(['year','State'])['Population'].sum().reset_index()#make data set with state i filtered for and taking the year and then get their population for each year you took 
i=0
for s in store_time_state['State'].unique():
    data = store_time_state[store_time_state['State'] ==s]#puts in data if it the info is for that specific state
    ax1.plot(data['year'], data['Population'], label=s, color=colors[i])# can also put all the states in for loop
    i += 1 #iterates color

ax1.set_xticks(range(2014,2019)) #unique didnt work since no 2017
ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax1.set_xlabel('year')
ax1.set_ylabel('Population')
ax1.set_title('Population Over Time')
ax1.ticklabel_format(style='plain',axis='y')

ny_cities = cities[cities['State'] == 'New York']
my_cities = ['Yonkers', 'Albany', 'New York', 'Utica', 'Syracuse']
ny_cities = ny_cities[ny_cities['City'].isin(my_cities)]
ny_2014 = ny_cities[ny_cities['year'] == 2014]
ny_2018 = ny_cities[ny_cities['year'] == 2018]
ny_2014 = ny_2014[['City','Population']]
ny_2018 = ny_2018[['City','Population']]
x = np.arange(len(my_cities))
width = 0.5
ax2.bar(x-width/2, ny_2014['Population'], width, label='2014')
ax2.bar(x+width/2, ny_2018['Population'], width, label='2018')
ax2.set_xticks(x, my_cities, rotation=45)
ax2.set_xlabel("City")
ax2.set_ylabel("Population")
ax2.set_title("NY Cities Populations in 2014 vs 2018")
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.ticklabel_format(style='plain',axis='y')

plt.tight_layout()
plt.show()
```
